## Zadanie 3
Poniżej zdefiniowałem funkcję f(x)=x^2 + 5 oraz sporządziłem jej wykresy dla różnych przedziałów.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def fun(x): # utworzenie funkcji
    return x**2+5

arg1 = np.linspace(-1,1,1000) # stworzenie trzech różnych przedziałów
arg2 = np.linspace(-6,6,1000)
arg3 = np.linspace(0,5,1000)

list1= [arg1,arg2,arg3]

for el in list1: # wykonanie prostej pentli
    plt.plot(el,fun(el),label='f(x)')
    plt.xlabel("Oś x")
    plt.ylabel("Oś y")
    plt.title("Wykres dla f(x)=x^2+5")
    plt.legend(title='Wartości:')
    plt.show()
```
