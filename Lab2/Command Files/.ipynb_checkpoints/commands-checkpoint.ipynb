```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

drinks_data = pd.read_csv('../Original Data/drinks.csv') # Read the data from csv
```

```python
drinks_data # Show dataframe
```

# Drinks data - countries sorted by beer servings

```python
drinks_copy = drinks_data[['country','beer_servings']] # Delated three columns
drinks_data_by_beer = drinks_copy.sort_values(by='beer_servings', ascending=False) # Sort data by descending
drinks_data_by_beer.index = np.arange(1, len(drinks_data_by_beer)+1) # Editing indexes
beer_10 = drinks_data_by_beer.head(10) # Create top 10 table
drinks_data_by_beer.to_csv('../Analysis Data/drinks_data_by_beer.csv',index = False) # Save edited data to csv in Analysis Data
drinks_data_by_beer # Show edited and saved data
```

# Drinks data - countries sorted by spirit servings

```python
drinks_copy_1 = drinks_data[['country','spirit_servings']] # Delated three columns
drinks_data_by_spirits = drinks_copy_1.sort_values(by='spirit_servings', ascending=False) # Sort data by descending
drinks_data_by_spirits.index = np.arange(1, len(drinks_data_by_spirits)+1) # Editing indexes
spirits_10 = drinks_data_by_spirits.head(10) # Create top 10 table
drinks_data_by_spirits.to_csv('../Analysis Data/drinks_data_by_spirits.csv',index = False) # Save edited data to csv in Analysis Data
drinks_data_by_spirits # Show edited and saved data
```
